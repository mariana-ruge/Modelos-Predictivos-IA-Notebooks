## Definir herramientas del Agente

```python
from llama_index.core.tools import BaseTool, FunctionTool
```

Operaciones matemáticas

```python
def multiply(a: int, b:int) -> int:
  "Multiplicar dos enteros y devolver el resultado"
  return a * b

#Usar Functool para las operaciones
multiply_tool = FunctionTool.from_defaults(fn=multiply)
```

```python
def add(a: int, b:int) -> int:
  "Sumar dos enteros"
  return a + b

#Usar Functool para las sumas
add_tool = FunctionTool.from_defaults(fn=add)
```

## Paso 5: Crear el Agente

```python
from llama_index.agent.openai import OpenAIAgent
from llama_index.llms.openai import OpenAI
from llama_index.core.prompts.system import SHAKESPEARE_WRITING_ASSISTANT
```

```python
#Definir el LLM (Large Language Model)
llm = OpenAI(model="gpt-3.5-turbo")

#Construir el objeto del agente
agent = OpenAIAgent.from_tools(
    [multiply_tool, add_tool],
    llm = llm,
    verbose=True
)
```

```python
#El pipeline llamará a un modelo preentrenado  en el modelo
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

Probar el modelo de la tonalidad  con la que se esta hablando

```python
classifier('We are happy to show you the Transformers Library')
```

```python
classifier('The food is not that cool')
```

Muestra que se esta dando un prompt positivo

Se puede entregar una lista de clasificadores

```python
#Usar un prompt como resultado
results = classifier(['We are happy to show you the Transformers Library', "We hope that you don't hate it"])

#Recorrer las emociones que se detecten
for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

## **Definir otro clasificador para el modelo**


Importar modelos para la clasificación

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

Identificar sentimientos en varios lenguajes (o idiomas), se requiere descargar los modelos y el tokenizador que se cargó

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

#Importación desde Hugging Face
tokenizer = AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment")

model = TFAutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment", from_pt= True)

classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
classifier("I am a good boy")
```

Ejecutar los tokens, tokenizar las frases a clasificar

```python
tf_batch = tokenizer(
    ["We are very happy to show you the transformers library", "We hope that you don't hate it"],
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="tf"
)
```

El padding se aplica automáticamente en el notebook y en el batch del modelo
